### Missing Values

We calculate the percentage of missing values in each column and show them in descending order.

```python
# Calculate percentage of missing values
mis_val_percentage = df.isnull().sum() / len(df) * 100
mis_val_percentage = mis_val_percentage[mis_val_percentage > 0]
mis_val_percentage = mis_val_percentage.sort_values(ascending=False)
mis_val_percentage = round(mis_val_percentage, 4)
# Convert to DataFrame
mis_val_percentage_df = pd.DataFrame(mis_val_percentage)
mis_val_percentage_df.reset_index(inplace=True)
mis_val_percentage_df.columns = ["Feature", "Missing Values (%)"]

mis_val_percentage_df
```

We can see that only 2 features or sensors have a significant amount of missing values.
- `sensor_15`: which has 100% missing values.
- `sensor_50`: which has 34.96% missing values.

It is evident that `sensor_15` is not useful for our analysis, and we can drop it.   
Although `sensor_50` has much less missing data, it is still significant compared to the rest of the features, which have less than 7%. Therefore, we decide to drop it as well.

```python
# Drop columns with more than 30% missing values
threshold = 50
missing_columns = mis_val_percentage[mis_val_percentage > threshold].index
df = df.drop(columns=missing_columns)
print(f"Columns with more than {threshold}% missing values dropped: {missing_columns.tolist()}")
```

# ** CAMBIAR **

For the rest of the features, we could fill the missing values with the mean of each feature. However, since the goal of this study is outlier detection, we will drop the rows with missing values to avoid introducing synthetic data that could affect the results of our analysis.

```python
# Impute missing values with mean for numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].mean())
```

### Correlation Status-Sensor

### Dimensionality Reduction

We will use **PCA** to reduce the dimensionality of the dataset.   
PCA (Principal Component Analysis) is a technique used to reduce the dimensionality of a dataset while preserving as much variance as possible. It does this by transforming the original features into a new set of uncorrelated features called principal components.

Moreover, by using PCA to transform the 52 sensors into just 2 components, we will be analyzing the data at the level of the overall functioning of the pump rather than focusing on the individual sensors.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = df.drop(['machine_status'], axis=1)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Add the original index to the PCA DataFrame
pca_df.index = df.index

# Add the machine status to the PCA DataFrame
pca_df['machine_status'] = df['machine_status'].values

mapping = {
    'NORMAL': 0,
    'RECOVERING': 1,
    'BROKEN': 2
}

# Copy the PCA DataFrame for sorting
pca_df_sorted = pca_df.copy()

# Map machine status to numerical values
pca_df_sorted['machine_status'] = pca_df_sorted['machine_status'].map(mapping)

# Sort the PCA DataFrame by machine status for better visualization
pca_df_sorted = pca_df_sorted.sort_values(by='machine_status', ascending=True)

# Reset label encoding for visualization
pca_df_sorted['machine_status'] = pca_df_sorted['machine_status'].map({0: 'NORMAL', 1: 'RECOVERING', 2: 'BROKEN'})

# Custom palette for machine status
custom_palette = {
    'NORMAL': '#1f77b4',  # Blue
    'RECOVERING': '#2ca02c',  # Green
    'BROKEN': '#d62728'  # Red
}

# Plot PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df_sorted, x='PC1', y='PC2', hue='machine_status', palette=custom_palette)
plt.title('PCA of Sensor Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Machine Status')
plt.show()
```
